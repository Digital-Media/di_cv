````python
import os
import cv2 # openCV
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
from plotly.express.colors import sample_colorscale
from plotly.subplots import make_subplots

!curl -o "cat.jpg" "https://raw.githubusercontent.com/Digital-Media/cv_data/main/example_images/cat.jpg" --silent
!curl -o "gogh.jpg" "https://raw.githubusercontent.com/Digital-Media/cv_data/main/example_images/gogh.jpg" --silent


def show_hist_stats(image: np.ndarray, use_cumulative: bool = False):
    """Function to create a histogram of an image and optionally display the stats.

    Args:
        image: The image to create the histogram for.
        use_cumulative: Whether to use the cumulative histogram or not.

    Returns:
        The figure object.
    """

    x = np.linspace(0, 1, 5)
    c = sample_colorscale('HSV', list(x))

    counts, bins = np.histogram(image.ravel(), bins=range(257))
    cumulative = np.cumsum(counts)

    fig = px.bar(
        x=bins[:-1],
        y=cumulative if use_cumulative else counts,
        labels={'x': 'pixel value', 'y': 'count'},
        color_discrete_sequence=['black'] * 256,
    )

    fig.update_layout(plot_bgcolor='white', margin=dict(t=0, b=0, r=0, l=0, pad=0))

    num_markers = 1000
    y_pos = -np.max(cumulative if use_cumulative else counts) * .05
    fig.add_traces([
        go.Scatter(
            x=np.linspace(0, 255, num_markers),
            y=[y_pos] * num_markers,
            mode='markers',
            marker={
                'color': np.linspace(0, 255, num_markers),
                'colorscale': 'gray', 'size': 10, 'symbol': 'square',
            },
        ),
    ])

    mean_value = np.mean(image)
    median_value = np.median(image)
    std_value = np.std(image)
    mode_value = np.argmax(counts)
    min_value = np.min(image)
    max_value = np.max(image)

    return fig


def show_image_and_hist(img, use_cumulative=False):
    """Function to display an image and its histogram side by side.

    Args:
        img: The image to display.
        use_cumulative: Whether to use the cumulative histogram or not.
    """
    fig = make_subplots(1, 2)
    fig.add_trace(
        go.Image(z=cv2.cvtColor(img, cv2.COLOR_GRAY2BGR), name="Image"), 1, 1,
    )
    traces = show_hist_stats(img, use_cumulative=use_cumulative).data
    for trace in traces:
        fig.add_trace(trace, 1, 2)
    fig.show()


def show_color_image(img):
    """Function to display a color image (BGR input).

    Args:
        img: The image to display (in BGR format).
    """
    fig = px.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    fig.update_layout(margin=dict(t=0, b=0, r=0, l=0, pad=0))
    fig.show()


def show_color_images_row(images, titles):
    """Function to display multiple color images (BGR) in a row.

    Args:
        images: List of images in BGR format.
        titles: List of titles for each image.
    """
    fig = make_subplots(1, len(images), subplot_titles=titles)
    for i, img in enumerate(images):
        fig.add_trace(
            go.Image(z=cv2.cvtColor(img, cv2.COLOR_BGR2RGB), name=titles[i]),
            1, i + 1,
        )
    fig.update_layout(height=400)
    fig.show()


color_img = cv2.imread("gogh.jpg")
show_color_image(color_img)
````
